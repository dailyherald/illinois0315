```python
import sys
import pandas as pd
import numpy as np
import time
from datetime import datetime
#from apscheduler.scheduler import Scheduler


import elex as elex
from elex.api import Election

print("ready")
```

```python
# NOTE this is set up for tests, not live
# also, to use AP's API you need an account with a key
election = Election(electiondate='2016-03-15', testresults=False, liveresults=True, is_test=False)
df=pd.read_csv('ILcands_prez.csv', dtype=object)

# this is the list of results using serialize
races = pd.DataFrame([f.serialize() for f in election.results])

allraces = races[['unique_id','first', 'last', 'officename', 'party', 'level', 'statename' , 'reportingunitname', 'precinctsreportingpct', 'precinctsreporting', 'precinctstotal', 'votecount', 'votepct','winner']]
racecols2 = pd.DataFrame( allraces.loc[ (allraces["officename"] == "President") ] )
racecols1 = pd.DataFrame( racecols2.loc[ (racecols2["level"] == "state" ) ] )
racecols = pd.DataFrame( racecols1.loc[ ( racecols1['unique_id'].isin(['polid-1746','polid-1445','polid-8639','polid-61815','polid-53044','polid-36679' ]) ) ] )
racecols.sort_values(['statename', 'party', 'last'], ascending=[True, True, False], inplace=True)
#racecols['precinctsreportingpct'] = pd.Series([round(val, 1) for val in racecols['precinctsreportingpct']], index = racecols.index)
#racecols['votepct'] = pd.Series([round(val * 100, 1) for val in racecols['votepct']], index = racecols.index)

# First state - Illinois
df_Illinois = pd.DataFrame( racecols.loc[ (racecols["statename"] == "Illinois" ) ] )
df_Illinois = df_Illinois[['unique_id', 'precinctsreportingpct', 'votepct']]
left = df
right = df_Illinois
df = pd.merge(left, right, how='left', on='unique_id', suffixes=('', '_Illinois'))
# second state - Ohio
df_Ohio = pd.DataFrame( racecols.loc[ (racecols["statename"] == "Ohio" ) ] )
df_Ohio = df_Ohio[['unique_id', 'precinctsreportingpct', 'votepct']]
left = df
right = df_Ohio
df = pd.merge(left, right, how='left', on='unique_id', suffixes=('', '_Ohio'))
# third state - North Carolina
df_NC = pd.DataFrame( racecols.loc[ (racecols["statename"] == "North Carolina" ) ] )
df_NC = df_NC[['unique_id', 'precinctsreportingpct', 'votepct']]
left = df
right = df_NC
df = pd.merge(left, right, how='left', on='unique_id', suffixes=('', '_NC'))
# fourth state - Missouri
df_Missouri = pd.DataFrame( racecols.loc[ (racecols["statename"] == "Missouri" ) ] )
df_Missouri = df_Missouri[['unique_id', 'precinctsreportingpct', 'votepct']]
left = df
right = df_Missouri
df = pd.merge(left, right, how='left', on='unique_id', suffixes=('', '_Missouri'))
# fifth state - Florida
df_Florida = pd.DataFrame( racecols.loc[ (racecols["statename"] == "Florida" ) ] )
df_Florida = df_Florida[['unique_id', 'precinctsreportingpct', 'votepct']]
left = df
right = df_Florida
df = pd.merge(left, right, how='left', on='unique_id', suffixes=('', '_Florida'))

df.to_csv('csv/fiveStates' + str( datetime.now() ) + '.csv', index=False, encoding="utf-8")
df.head(5)
```
